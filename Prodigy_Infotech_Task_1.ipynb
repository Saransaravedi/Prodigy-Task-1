Data Visualization :

```python
sns.scatterplot(data)
```

```python
siz = (35, 10)
fig, ax = plt.subplots(figsize=siz)
sns.barplot(ax=ax,data=data)
```

```python
years = data.columns[4:]
years
```

```python
years = data.columns[4:]
values = data[years].sum()
plt.figure (figsize=(30,20))
plt.barh(years,values,edgecolor='blue')
plt.xlabel('Total Values')
plt.ylabel('Year', size=20)
plt.title('Total Values per Year', size=20)
plt.show()
```

```python
for i in years:
  fig = plt.figure(figsize=(5,5))
  plt.hist(data[i],color ='pink',bins=10)
  plt.xlabel(i)
  plt.show()
```

```python
sns.histplot(data, kde=False, bins=5, color='skyblue')
plt.title('Total Values Per Year')
plt.xlabel('Values')
plt.ylabel('years')
plt.show()
```
